# Map Reduce &Filter

```python
#map
```

```python
list1 = [1,2,3,4]
sq=list(map(lambda sq:sq**2,list1))
sq
```

```python
sq=list(map(lambda sq:sq**2,[1,2,3,4]))
sq
```

```python
sq=list(map(lambda x:x**2,[1,2,3,4]))
sq
```

```python
odd_even = list((map(lambda x:'Even' if x%2==0 else 'Odd',[1,2,3,4]))
odd_even
```
